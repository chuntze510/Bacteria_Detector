```python
#generate flip img
label=0
for k in range(0,26):
    imgclass=1+k
    imgname='0 ('+str(imgclass)+').jpg'
    img = cv2.imread(imgname)
    label=label+1
    filename=('0_'+str(label)+'.jpg')
    img_flip = cv2.flip(img, 1)
    print(img_flip.shape)
    cv2.imwrite(filename,img_flip)      
    print('已經存檔:第',label,'張',filename)
print('裁減完成!!!!')
```

```python
# 讀取圖檔 自動把兩千成兩千裁剪成六份
label=0
for k in range(0,52):
    imgclass=1+k
    imgname='0 ('+str(imgclass)+').jpg'
    img = cv2.imread(imgname)
    for i in range (0,6):
            label=label+1
            filename=('0_'+str(label)+'.jpg')
            x = 0+400*i
            y = 0
            # 裁切區域的長度與寬度
            w = 4000
            h = 4000
            # 裁切圖片
            crop_img = img[y:y+h, x:x+w]
            crop_img=cv2.resize(crop_img,(416,416))
            print(crop_img.shape)
            cv2.imwrite(filename,crop_img)      
            print('已經存檔:第',label,'張',filename)
print('裁減完成!!!!')
```
